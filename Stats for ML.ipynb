## Make a bar graph for manufacturer type

```python
# Group the data by manufacturer and calculate the count for each manufacturer type 
manufacturer = df.groupby('Manufacturer')['Manufacturer'].count()

# Sort the values in descending order
manufacturer = manufacturer.sort_values(ascending=False)

# Bar graph for manufacturer type
plt.figure(figsize=(10, 6))
manufacturer.plot(kind='bar', color='red', alpha=0.5)
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.title('Count of Manufacturer Type')
plt.xticks(rotation=90)
plt.show()
```

## Make a pie chart for showing vehicle type by percentage

```python
# Group the data by vehicle type and calculate the percentage of each vehicle type
vehicle_type_percentage = df.groupby('Vehicle_type')['Vehicle_type'].count() / len(df['Vehicle_type']) * 100
vehicle_type_percentage = round(vehicle_type_percentage, 2)   
vehicle_type_percentage

# Pie chart of vehicle type by percentage
plt.figure(figsize=(4, 4))
vehicle_type_percentage.plot(kind='pie', autopct='%1.2f%%')                        
plt.title('Vehicle Type by Percentage')
plt.ylabel('')
plt.show()
```

## Calculate the variance and standard deviation for numerical columns such as Sales_in_thousands, __year_resale_value, Price_in_thousands, Engine_size, Horsepower

```python
# Iterate over all columns and compute variance and standard deviation of numerical columns
for c in df.columns:
    if(df[c].dtype == 'float64'):
        print(df[c].name)
        print("---------------------------")
        print("Variance:\t\t", round(np.var(df[c]), 2))
        print("Standard Deviation:\t", round(np.std(df[c]), 2))
        print("\n\n")
```

## For the same columns, calculate IQR and identify outliers in these columns

```python
# Iterate over all columns, calculate IQR and identify outliers for numerical columns 
for c in df.columns:
    if(df[c].dtype == 'float64'):
        print(df[c].name)
        print("---------------------------")
        Q1 = df[c].quantile(0.25)
        Q3 = df[c].quantile(0.75)
        IQR = Q3 - Q1
        print("IQR:\t\t", IQR, "\n")
        outliers = df[(df[c] < Q1 - 1.5 * IQR) | (df[c] > Q3 + 1.5 * IQR)][c]
        print("Outliers:\t\t", outliers)
        print("\n\n")
```

## Make Histograms for all numerical columns and identify the distribution of type by calculating skewness.


*Note: If the skewness is 0, it indicates a perfectly symmetrical distribution.* \
*If the skewness is negative, it means the distribution is skewed to the left (long left tail).* \
*If the skewness is positive, it means the distribution is skewed to the right (long right tail).*

```python
# Histograms of all numerical columns
# Select numeric columns only
numeric_columns = df.select_dtypes(include=['number'])

# Get the number of numeric columns
num_cols = len(numeric_columns.columns)

# Calculate the number of rows and columns for the subplots
num_rows = (num_cols // 2) + (num_cols % 2)  # Ensures that odd numbers have an extra row
num_cols_per_row = 2

# Create subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols_per_row, figsize=(20, 16))

# Flatten the 2D array of subplots into a 1D array
axes = axes.flatten()

# Plot data on each subplot
for i, column in enumerate(numeric_columns.columns):
    axes[i].hist(df[column], bins=20, color='blue', edgecolor='black')
    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate skewness for all numeric columns
for c in df.columns:
    if(df[c].dtype == 'float64'):
        print(df[c].name)
        print("---------------------------")
        skewness = round(skew(df[c]), 2)
        print("Skewness:", skewness)
        if(skewness == 0):
            print("The distribution is perfectly symmetrical.")
        elif(skewness > 0):
            print("The distribution is skewed to the right.")
        elif(skewness < 0):
            print("The distribution is skewed to the left.")
        print("\n\n")
```
